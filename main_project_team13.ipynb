# Exploratory Data Analysis

 ## Univariate Analysis

```python
plt.figure(figsize=(18,6)) 
plt.subplot(121)
sns.kdeplot(data=data, x="fy_declared",fill=True)
plt.xlabel("fy_declared")
plt.tight_layout(pad=0.5)

plt.subplot(122)
sns.kdeplot(data=data, x="disaster_number",fill=True)
plt.xlabel("disaster_number")
plt.tight_layout(pad=0.5)
```

```python
ax=sns.countplot(x = 'declaration_type',data = data,color="salmon")
plt.title(" Declared Disasters Count Since 1953")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center")
```

```python
plt.figure(figsize=(20,10)) 
plt.xticks(rotation=45)
ax = sns.countplot(x="incident_type", data= data, palette="plasma")
ax.set_title('Incident type Count Since 1953', fontsize=20)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center",size=18)
plt.show()
```

```python
plt.subplot(141)
data['ih_program_declared'].plot.kde(figsize=(15,6))
plt.xlabel("ih_program_declared")
plt.tight_layout(pad=0.5)
plt.subplot(142)
data['ia_program_declared'].plot.kde()
plt.xlabel("ia_program_declared")
plt.tight_layout(pad=0.5)
plt.subplot(143)
data['pa_program_declared'].plot.kde()
plt.xlabel("pa_program_declared")
plt.tight_layout(pad=0.5)
plt.subplot(144)
data['hm_program_declared'].plot.kde()
plt.xlabel("hm_program_declared")
plt.tight_layout(pad=0.5)



```

## Bivariate Analysis

```python
plt.figure(figsize=(20,15))
plt.xticks(rotation=90)
sns.boxplot(y = "fy_declared", x = "incident_type", data=data)
```

```python
sns.pairplot(
    data,
    x_vars=[ 'declaration_type','fy_declared'],
    y_vars=['incident_type'],height=5   
)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4,figsize=(20,6)) # create 1x4 array of subplots
sns.boxplot(y = "disaster_number", x = "ia_program_declared", data=data, ax=axes[0]) # add boxplot to 1st subplot
sns.boxplot(y = "disaster_number", x = "pa_program_declared", data=data, ax=axes[1])
sns.boxplot(y = "disaster_number", x = "ih_program_declared", data=data, ax=axes[2]) 
sns.boxplot(y = "disaster_number", x = "hm_program_declared", data=data, ax=axes[3])
```

```python
# Identifying the outliers
Q1 = np.percentile(data['fy_declared'],25,interpolation='midpoint')
Q3 = np.percentile(data['fy_declared'],75,interpolation='midpoint')
IQR = Q3-Q1
low_limit = Q1-1.5*IQR     
up_limit = Q3 + 1.5*IQR
outlier = []
for x in data['fy_declared']:
    if (x > up_limit) or (x < low_limit):
        outlier.append(x)       
```
